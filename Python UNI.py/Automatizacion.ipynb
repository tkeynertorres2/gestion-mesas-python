```python
data= yfinance.Ticker("AAPL").history("6mo")
cierre=data.Close
```

## HACER ANALISIS SOLICITADO POR EL GESTOR 
* cotizacion max
* cotizacion min
* cotizacion media 

```python
maxima=(round(cierre.max(),2))
minima=(round(cierre.min(),2))
media=(round(cierre.mean(),2))
print (maxima)
print (minima)
print (media)
```
