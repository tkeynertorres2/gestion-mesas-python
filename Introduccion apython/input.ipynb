# Entrada de datos con la función input()

En la programación, es muy común requerir que el usuario ingrese datos para la ejecución de un programa.

Con el lenguaje de programación python, podemos recibir datos de un usuario a través de la función input().

Esta función recibe como parámetro una cadena de texto ó string, que será mostrada al usuario, normalmente con el fin de validar el tipo de dato que el programa "espera" recibir.


```python
numero_1= int(input("Ingrese el primer numero"))
numero_2= int(input("Ingrese el segundo numero"))
print (numero_1)
print (numero_2)
print (numero_1+numero_2)
```
