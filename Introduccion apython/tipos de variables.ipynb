# Qué son las variables?
Las variables son unos de los recursos más básicos de los lenguajes de programación.

A través de las variables, podemos almacenar datos en memoria para acceder a ellos a lo largo de nuestro código/programa, cuando sea necesario.

Declaración y asignación

Algunas reglas deben ser seguidas al momento de elegir el nombre de una variable.

* Debe comenzar con una letra o un underscore (_).

* No puede comenzar con un número.

* Los nombres de las variables son case sensitive (edad es diferente a Edad).

* No se puede utilizar una palabra reservada del lenguaje.

* No pueden contener espacios.

```python
edad=25
print(type(edad))
```

```python
altura=1.74
print(type(altura))
```

```python
ciudad_origen = "villavicencio"
print(type(ciudad_origen))
```

```python
usuario_gold= True
print(type(usuario_gold))
```
